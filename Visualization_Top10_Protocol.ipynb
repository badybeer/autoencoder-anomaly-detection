```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Natanie datasetu
file_path = 'Datasety/split_1.csv'
data = pd.read_csv(file_path)

# Poet vskytov pre kad aplikciu
application_counts = data['application_name'].value_counts()

# Zskanie top 10 aplikci
top_10_applications = application_counts.head(10)

# Vykreslenie grafu
plt.figure(figsize=(10, 8))
top_10_applications.plot(kind='bar')
plt.title('Distribcia Top 10 Protokolov v Datasete', fontsize=18)
plt.xlabel('Protokol', fontsize=16)
plt.ylabel('Poet', fontsize=16)
plt.grid(True, which='major', linestyle='--', linewidth=0.75, color='grey')
plt.xticks(rotation=45, ha='right', fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout()

# Uloenie grafu do sboru s nzvom 'bidirectional_packets_plot.png' v tlaovej kvalite (300 DPI)
plt.savefig('Figure\Distribcia Top 10 Protokolov v Datasete).png', dpi=300)

# Zobrazenie grafu
plt.show()
```
